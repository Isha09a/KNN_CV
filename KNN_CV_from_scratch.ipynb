```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module='bs4')
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

KNN from scratch

```python
# Loading the dataset
df=pd.read_csv('/content/drive/MyDrive/Real estate.csv')
```

```python
# Dropping the index column
df.drop('No',axis=1,inplace=True)
```

```python
# Checking for null values
df.isna().sum()
```

```python
# checking for duplicated values
df.duplicated().sum()
```

```python
# separating label and X
df_num=df.drop('Y house price of unit area',axis=1)
y=df[['Y house price of unit area']]
```

```python
# Standardizing the data
ss=StandardScaler()
X=ss.fit_transform(df_num)
X=pd.DataFrame(X,columns=df_num.columns)
```

```python
X.shape,y.shape
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Function for knn-regression
def KnnReg(X_train, X_test, y_train, y_test,K=3):

    train=pd.concat([X_train,y_train],axis=1)
    test=pd.concat([X_test,y_test],axis=1)
    train.shape,test.shape
    test=test.set_index(keys=np.arange(0,len(test)))
    train=train.set_index(keys=np.arange(0,len(train)))
    dist=[]
    t=0
    k=K

    # finding the distance of test record from train records
    for i in range(test.shape[0]):
        distances_for_i = []
        for j in range(train.shape[0]):
            total=0
            for c in range(train.shape[1]-1):
                total+=(test.iloc[i,c]-train.iloc[j,c])**2
            d=np.sqrt(total)
            distances_for_i.append(d)
        dist.append(distances_for_i)

    # selecting y having  k min distances

    predictions = []
    for i in range(len(dist)):
        distances_for_i = dist[i]
        indices = np.argsort(distances_for_i)[:k]
        near_y_labels = [train.iloc[j, -1] for j in indices]
        y_hat = np.mean(near_y_labels)
        predictions.append(y_hat)

    # concating predicted value into the test data
    X_test['y_pred']=predictions
    return X_test
```

```python
KnnReg(X_train, X_test, y_train, y_test,4)
```

Cross Validation from scratch

```python
# Cross-validation code:
def Cross_Validation(cv=5,df=df,k=3):

    # Shuffling the dataset for random selection
    shuffled_df = df.sample(frac=1, random_state=42)
    shuffled_df = shuffled_df.reset_index(drop=True)

    len_df = df.shape[0]
    train_mape_list = []
    test_mape_list = []

    for i in range(cv):
        start = i * len(df) // cv
        end = (i + 1) * len(df) // cv
        train = pd.concat([shuffled_df.iloc[:start,:-1], shuffled_df.iloc[end:,:-1]])
        test = df.iloc[start:end,:-1]
        train_y = pd.concat([shuffled_df.iloc[:start,-1], shuffled_df.iloc[end:,-1]])
        test_y = shuffled_df.iloc[start:end,-1]

        # Implement model
        test_new = KnnReg(train,test,train_y,test_y,3)

        # Getting MAPE score
        test_mape = mean_absolute_percentage_error(test_y,test_new['y_pred'])
        test_mape_list.append(test_mape)
    return(test_mape_list)
```
